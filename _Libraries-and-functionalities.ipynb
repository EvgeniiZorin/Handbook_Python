# _Introduction

Module: a single `.py` file containing functions, classes, and variables. E.g. `math`, `random`, `datetime`

Library: a collection of related modules and packages, e.g. `numpy`, `pandas`, `matplotlib`

**How is a module imported?**

Any way you import a module, the `math` module was loaded into memory and referenced in `sys.modules`. 
- Importing: `import math` or `from math import sqrt` are equally as efficient. 
- Calling: `math.sqrt(2)` is technically slower than `sqrt(2)` because requires to find `sqrt` in `math`'s namespace, but practically, no difference

**Import variants**

Let's say you have a module `module1.py`. In there, you can import modules and functions in different ways:

```py
import sys

# Variant 1
import math
# math is in sys.modules as `math`
# math is in global namespace `globals()`


# Variant 2
import math as r_math
# math is in sys.modules as `math`
# math is in global namespace `globals()` as `r_math`


# Variant 3
from math import sqrt
# math is in sys.modules as `math`
# in global namespace `globals()`, you have the function listed as `sqrt`, but not math
# (if any symbols already exists in module1's namespace, replace their reference)


# Variant 4
from math import sqrt as r_sqrt
# math is in sys.modules as `math`
# in global namespace `globals()`, you have the sqrt function listed as `r_sqrt`
# (if any symbols already exists in module1's namespace, replace their reference)


# Variant 5
from math import *
# math is in sys.modules as `math`
# in global namespace `globals()`, you now have all the functions from `math` module - `pi`, `sin`, etc...
# (if any symbols already exists in module1's namespace, replace their reference)
```

## Reloading modules

````python
import os

def create_module_file(module_name, **kwargs):
    """
    Create a module file named <module_name>.py
    Module has a single function (print_values) that will
    print out the supplied (stringified) kwargs.
    """
    module_file_name = f"{module_name}.py"
    module_rel_file_path = module_file_name
    module_abs_file_path = os.path.abspath(module_rel_file_path)
    
    with open(module_abs_file_path, 'w') as f:
        f.write(f"# {module_name}.py\n\n")
        f.write(f"print('running {module_file_name}...')\n\n")
        f.write(f"def print_values():\n")
        for key, value in kwargs.items():
            f.write(f"\tprint('{str(key)}', '{str(value)}')\n")

create_module_file('test', k1=10, k2='python')

import test
````

## Package

Use case:
- Code organisation: 
  - As module developer, you can group related functions into a module
  - As module user, you want to import easier, e.g. `from module1 import func1` instead of being very specific like `from module1.submodule1.import func1`
  - this modular code organisation makes our code easier to write, test and debug, read/understand, and document (just like a book broken down into chapter, paragraphs, etc.)
- Hides inner implementation from the user, making it easier to use

- Packages are modules, but modules are not necessarily packages
- packages represent a hierarchy of modules / packages
- package is simply a module that can contain other modules / packages
- package name: directory name
- contains `__init__.py`
- packages (as opposed to modules) have a `__path__` property

To define a package in our file system, we must:
- create a directory whose name will be the package name
- create a file `__init__.py` inside that directory; this file tells Python that the directory is a package as opposed to a standard directory

Example of a package `pack1`:

```txt
app/
    pack1/
        __init__.py
        module1.py
        module2.py
```

Then `__init__.py` has the following code:

```py
import pack1.module1
import pack1.module2
```

Then we import the package:
```py
import pack1
```

- The code for `pack1` is in `__init__.py`
- that code is loaded, executed and cached in `sys.modules` with a key of `pack1`
- the symbol `pack1` is added to our namespace referencing the same object
- `module1` and `module2` are imported and added to `sys.modules`

# API

**API** - Application Programming Interface. Web service that provides an interface to applications to manipulate and retrieve information. 

**API endpoint** - the point of entry in a communication channel when two systems are interacting. It refers to touchpoints of the communication between an API and a server. Examples include `/hello`, `/get-item`.

`base_url/endpoint`, for example:
- `localhost/get-item`
- `facebook.com/home`, where `facebook.com` - base url, `/home` - endpoint

HTTP methods (requests):
- GET: this endpoint returns information to you
- POST: you send information to this endpoint; will be creating something new
- PUT: update database information
- DELETE: delete database information

Parameters:
- **Path parameters** are passed as path, like this: `http://127.0.0.1:8000/get-item/3`
- **Query parameters** come after question mark in URL: `http://127.0.0.1:8000/get-by-name?name=Milk` or like this `http://127.0.0.1:8000/get-by-name3?test=3&name=3`

## Combine pdfs to one pdf

````python
import os

from pypdf import PdfWriter
import fitz

def merge_pdfs(input_folder, output_filename):
    """
    Merges all PDF files in a given folder into a single PDF.

    Args:
        input_folder (str): The path to the folder containing the PDF files.
        output_filename (str): The name of the output merged PDF file.
    """
    merger = PdfWriter()
    pdf_files = []

    # Collect all PDF files in the specified folder
    for root, _, filenames in os.walk(input_folder):
        for filename in filenames:
            if filename.lower().endswith('.pdf'):
                pdf_files.append(os.path.join(root, filename))

    # Sort the files to ensure a consistent merging order (optional)
    pdf_files.sort()

    # Append each PDF file to the merger object
    for pdf_file in pdf_files:
        merger.append(pdf_file)

    # Write the merged PDF to the output file
    with open(output_filename, "wb") as output_file:
        merger.write(output_file)

    merger.close()
    print(f"Successfully merged {len(pdf_files)} PDF files into {output_filename}")


# # Example usage:
# # Create a subfolder named 'pdfs_to_merge' and place your PDF files inside
# # The merged PDF will be saved in the same directory as the script
# input_directory = "data/pdfs_passport_res-perm"
# output_pdf_name = "data/output/pdfs_passport_res-perm.pdf"

# # Ensure the input directory exists for the example to work
# if not os.path.exists(input_directory):
#     os.makedirs(input_directory)
#     print(f"Created directory: {input_directory}. Please place your PDF files inside.")
# else:
#     merge_pdfs(input_directory, output_pdf_name)
````

````python
import fitz  # PyMuPDF
from PIL import Image
import io
import os

def compress_pdf(input_path, output_path, image_quality=60, skip_small_images=True):
    """
    Safely compress PDF images using Pillow + PyMuPDF.
    
    Args:
        input_path (str): Path to source PDF.
        output_path (str): Path to save compressed PDF.
        image_quality (int): JPEG quality (0–100).
        skip_small_images (bool): Skip recompressing very small images (e.g., icons).
    """
    src = fitz.open(input_path)
    dst = fitz.open()  # new output PDF

    for page_index in range(len(src)):
        page = src.load_page(page_index)
        pix = page.get_pixmap(dpi=150)  # render page to raster
        img = Image.open(io.BytesIO(pix.tobytes("ppm")))

        # Save as JPEG-compressed version of the page
        img_buffer = io.BytesIO()
        img.save(img_buffer, format="JPEG", quality=image_quality)
        img_buffer.seek(0)

        # Add new page with the compressed image
        rect = page.rect
        new_page = dst.new_page(width=rect.width, height=rect.height)
        new_page.insert_image(rect, stream=img_buffer.getvalue())

    # Save compressed PDF
    dst.save(output_path, garbage=4, deflate=True)
    dst.close()
    src.close()

    # Size comparison
    orig_size = os.path.getsize(input_path) / 1024
    new_size = os.path.getsize(output_path) / 1024
    print(f"✅ Compressed PDF saved: {output_path}")
    print(f"📦 Size reduced: {orig_size:.1f} KB → {new_size:.1f} KB ({100*(1 - new_size/orig_size):.1f}% smaller)")



# compress_pdf(
#     'data/241021_birth certificate - translation ESP - updated corrected MAXIM.pdf',
#     'data/output/241021_birth certificate - translation ESP - updated corrected MAXIM_reduced.pdf',
#     image_quality=25
# )
````

````python
from pypdf import PdfWriter

def reduce_filesize_pdf(FILENAME):
    """
    Given a path of a pdf (without the `.pdf` extension), reduce it's size. 
    
    Works especially well with very large pdf sizes. 
    
    More info: https://pypdf.readthedocs.io/en/latest/user/file-size.html
    """
    print(f'... Processing {FILENAME}.pdf')
    writer = PdfWriter(clone_from = f'example_datasets/{FILENAME}.pdf')

    for page in writer.pages:
        # Lossless compression (try this if the image quality reduction doesn't work)
        # page.compress_content_streams(level=9)
        # Reduce image quality
        for img in page.images:
            img.replace(img.image, quality=80)

    with open(f'example_datasets/{FILENAME}_compressed.pdf', 'wb') as f:
        writer.write(f)

    print('Done.')
    return None

for i in ['Dark_forest', 'joinedPDF']:
    reduce_filesize_pdf(i)
````

# Threading

In Python, there is no real multi-threading because of the global interpreter lock. 

Concurrency - two or more tasks are in progress at any point in time, rather than sequentially completing the tasks. IOW, concurrent program execution is the opposite of sequential program execution. 

Concurrency can occur in multiple ways:
- **Asynchrony**: processing just refers to the fact that some tasks are performed in the background without you having to explicitly wait for the results. You are informed of the results at a later point in time.
- **Parallelism**: Two tasks are being executed at the same time (e.g. on separate cores). It’s a bit like singing in the shower: you’re actually doing two things at exactly the same time.
- **Multithreading**: 



## Ex1: simple

````python
### An older way to do threading
### Method 1
import threading
import time

start = time.perf_counter()

def do_something():
    print('sleeping 1 second ...')
    time.sleep(1)
    print('Done sleeping.')

t1 = threading.Thread(target=do_something)
t2 = threading.Thread(target=do_something)
t3 = threading.Thread(target=do_something)

t1.start()
t2.start()
t3.start()

### This makes sure that these two threads finish first, 
### before moving onto the next task of printing (finish-start)
t1.join()
t2.join()
t3.join()


finish = time.perf_counter()
print(finish-start)

### In the end, it would have taken 3 seconds running sequentially, 
### but as we run the three tasks concurrently, it will only take a second
````

````python
### Method 2

import threading
import time

start = time.perf_counter()

def do_something():
    print('sleeping 1 second ...')
    time.sleep(1)
    print('Done sleeping.')

threads = []

for _ in range(10):
    t = threading.Thread(target=do_something)
    # If you want to pass arguments, you can do like this: target=do_something, args=[1.5]
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()


finish = time.perf_counter()
print(finish-start)
````

````python
### Method 3
##3 now using another module

import concurrent.futures
import time

start = time.perf_counter()

def do_something():
    print('sleeping 1 second ...')
    time.sleep(1)
    return 'Done sleeping.'

with concurrent.futures.ThreadPoolExecutor() as executor:
    ### Run one function once
    # # f1 = executor.submit(do_something, 1) # if the function takes an argument
    # f1 = executor.submit(do_something)
    # print(f1.result())
    ### Run the same function multiple times
    results = [executor.submit(do_something) for _ in range(10)]
    for f in concurrent.futures.as_completed(results):
        print(f.results())

# threads = []

# for _ in range(10):
#     t = threading.Thread(target=do_something)
#     # If you want to pass arguments, you can do like this: target=do_something, args=[1.5]
#     t.start()
#     threads.append(t)

# for thread in threads:
#     thread.join()


finish = time.perf_counter()
print(finish-start)
````

````python
import time

def time_it(fn, # the function fn to call 
            *args, # positional arguments to use in the function fn
            rep = 1, # how many times you want to repeat calling the function fn
            mean = True, # calculate mean or different / sum?
            **kwargs, # keyword arguments to use in the function fn
            ) -> None:
    """
    Runs a passed in object (function) with the specified arguments
    a specified amount of times, and gives an average time to run it.
    """
    start = time.perf_counter()
    for i in range(rep):
        fn(*args, **kwargs)
    end = time.perf_counter()
    return (end - start) / rep if mean else end - start

time_it(print, 
        1, 2, 3, 
        rep = 10,
        sep=' - ', 
        end=' ***\n'
        )
````

````python
def computer_powers_1(n, *, start = 1, end):
    # using a for loop
    results = []
    for i in range(start, end):
        results.append(n**i)
    return results

computer_powers_1(2, end = 5)

def computer_powers_2(n, *, start = 1, end):
    # using a list comprehension
    return [n**i for i in range(start, end)]

computer_powers_2(2, end = 5)

def computer_powers_3(n, *, start = 1, end):
    # using generators expression
    return (n**i for i in range(start, end))

list(computer_powers_3(2, end = 5))
````
